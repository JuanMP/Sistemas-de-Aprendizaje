```python
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from statsmodels.graphics.api import plot_corr
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

#Cargar datos (especificando la codificación utf-8)
df = pd.read_csv('Housing_Modified.csv', encoding='utf-8')

#Convertir campos binarios a campos booleanos numéricos
lb = preprocessing.LabelBinarizer()
for col in ['driveway', 'recroom', 'fullbase', 'gashw', 'airco', 'prefarea']:
    df[col] = lb.fit_transform(df[col])

#Crear variables dummy para 'stories'
df_stories = pd.get_dummies(df['stories'], prefix='stories', drop_first=True)

#Unir las variables dummy al dataframe principal
df = pd.concat([df, df_stories], axis=1)

#Eliminar la columna original 'stories'
del df['stories']

#Codificar las dummies como valores numéricos (1 = True, 0 = False)
df['stories_one'] = df['stories_one'].map({True: 1, False: 0})
df['stories_two'] = df['stories_two'].map({True: 1, False: 0})
df['stories_three'] = df['stories_three'].map({True: 1, False: 0})

#Definir la lista de columnas predictoras
preds = ['lotsize', 'bedrooms', 'bathrms', 'driveway', 'recroom', 'fullbase', 'gashw',
         'airco', 'garagepl', 'prefarea', 'stories_one', 'stories_two', 'stories_three']
X = df[preds]

#Límite de VIF
limite = 10

#Bucle para eliminar columnas con VIF mayor al límite
while True:
    vif = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    max_vif = max(vif)
   
    if max_vif > limite:
        maxloc = vif.index(max_vif)
        print(f"Eliminando '{X.columns[maxloc]}' con VIF = {max_vif:.2f}")
        X = X.drop(columns=[X.columns[maxloc]])
    else:
        break

#Mostrar VIF final de las columnas restantes
vif_final = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
for i, col in enumerate(X.columns):
    print(f"{col}: VIF = {vif_final[i]:.2f}")

#Calcular la matriz de correlaciones
corr = X.corr()

#Dibujar la matriz de correlaciones usando plot_corr() de Statsmodels
plot_corr(corr, xnames=list(corr.columns))
plt.show()
```

### EJERCICIO 14

Tenemos los datos de 'Grade_Set_2.csv', considera otro conjunto de datos de notas de estudiantes

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Carga datos
df = pd.read_csv('Grade_Set_2.csv')
print(df)

# Gráfico
df.plot(kind='scatter', x='Hours_Studied', y='Test_Grade', title='Grade vs Hours Studied')
plt.show() # added to show the graph.

# check the correlation between variables
print("Correlation Matrix: ")
print(df.corr()) # print the correlation matrix

# Create linear regression object
lr = LinearRegression()
x = df.Hours_Studied.values.reshape(-1, 1)  #independent variable, reshape is required
y = df.Test_Grade  # dependent variable

# Train the model using the training sets
lr.fit(x, y)

# plotting fitted line
plt.scatter(x, y, color='black')
plt.plot(x, lr.predict(x), color='blue', linewidth=3)
plt.title('Grade vs Hours Studied')
plt.ylabel('Test_Grade')
plt.xlabel('Hours_Studied')
plt.show() # added to show the graph

print("R Squared: ", r2_score(y, lr.predict(x)))
```

El análisis de correlación muestra el 97% de relación positiva entre las horas estudiadas y la nota obtenida y el 95% (R2) de variación puede ser explicada por las horas estudiadas. Observa que menos
de 4 horas de media estudiadas da una puntuación menor de 30 y cada 9 horas añade a la nota un nivel. Esto puede modelarse también con un regresor polinómico.

```python
lr = LinearRegression()
x= df.Hours_Studied # independent variable
y= df.Test_Grade # dependent variable
for deg in [1, 2, 3, 4, 5]:
 lr.fit(np.vander(x, deg + 1), y) # función vander() devuelve potencias del vector
 y_lr = lr.predict(np.vander(x, deg + 1))
 plt.plot(x, y_lr, label='degree ' + str(deg))
 plt.legend(loc=2)
 print (r2_score(y, y_lr))
 plt.plot(x, y, 'ok')
```
