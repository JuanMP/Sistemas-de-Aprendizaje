```python
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from statsmodels.graphics.api import plot_corr
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

#Cargamos los datos
df = pd.read_csv('Housing_Modified.csv', encoding='utf-8')

#Convertir campos binarios a campos booleanos numéricos
lb = preprocessing.LabelBinarizer()
for col in ['driveway', 'recroom', 'fullbase', 'gashw', 'airco', 'prefarea']:
    df[col] = lb.fit_transform(df[col])

#Creamos variables dummy para 'stories'
df_stories = pd.get_dummies(df['stories'], prefix='stories', drop_first=True)

#Unir las variables dummy al dataframe principal
df = pd.concat([df, df_stories], axis=1)

#Eliminamos la columna original 'stories'
del df['stories']

#Codificamos las dummies como valores numéricos (1 = True, 0 = False)
df['stories_one'] = df['stories_one'].map({True: 1, False: 0})
df['stories_two'] = df['stories_two'].map({True: 1, False: 0})
df['stories_three'] = df['stories_three'].map({True: 1, False: 0})

#Definimos la lista de columnas predictoras
preds = ['lotsize', 'bedrooms', 'bathrms', 'driveway', 'recroom', 'fullbase', 'gashw',
         'airco', 'garagepl', 'prefarea', 'stories_one', 'stories_two', 'stories_three']
X = df[preds]

#Límite de VIF
limite = 10

#Bucle para eliminar columnas con VIF mayor al límite
while True:
    vif = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    max_vif = max(vif)
   
    if max_vif > limite:
        maxloc = vif.index(max_vif)
        print(f"Eliminando '{X.columns[maxloc]}' con VIF = {max_vif:.2f}")
        X = X.drop(columns=[X.columns[maxloc]])
    else:
        break

#Mostrar VIF final de las columnas restantes
vif_final = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
for i, col in enumerate(X.columns):
    print(f"{col}: VIF = {vif_final[i]:.2f}")

#Cálculo de la matriz de correlaciones
corr = X.corr()

#Dibujamos la matriz de correlaciones usando plot_corr() de Statsmodels
plot_corr(corr, xnames=list(corr.columns))
plt.show()
```

### EJERCICIO 11

Ejecuta el ejemplo 21 y el ejemplo 22 aplicado al modelo del ejercicio anterior. Indica qué datos son outliers. Ahora aplica alguno de los algoritmos básicos (o el basado en z-scores o el basado en IQR) y comprueba si detectan los mismos outliers>

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.graphics.regressionplots import plot_leverage_resid2

#Cargamos datos
df = pd.read_csv('Housing_Modified.csv')

#Solo columnas numéricas y rellenar valores NaN con la media
df = df.select_dtypes(include=[np.number])
df.fillna(df.mean(), inplace=True)

#Definir las variables independientes (X) y dependientes (y)
X = df.drop(columns=['price'])
X = sm.add_constant(X)  # Añadir constante
y = df['price']

#Eliminar valores infinitos y filas con NaN
X = X.replace([np.inf, -np.inf], np.nan).dropna()
y = y.loc[X.index]

#Comprobación de VIF y eliminación de variables con alto VIF
limite = 10
while True:
    vif = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    max_vif = max(vif)
    if max_vif > limite:
        max_index = vif.index(max_vif)
        print(f"Eliminando '{X.columns[max_index]}' con VIF={max_vif:.2f}")
        X = X.drop(columns=[X.columns[max_index]])
    else:
        break

#Entrenamos el modelo
modelo = sm.OLS(y, X).fit()
print(modelo.summary())

#Mostrar gráfico de leverage vs residuos
fig, ax = plt.subplots(figsize=(8, 6))
plot_leverage_resid2(modelo, ax=ax)
plt.title('Gráfico de leverage vs residuos')
plt.show()
```

### EJERCICIO 12

Modifica el ejemplo 28 para calcular también un modelo lineal a los mismos datos y
muestra el R2 del modelo lineal y del modelo polinómico con los datos de entrenamiento

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Generar datos
n = 100 # Número de datos
np.random.seed(0) # Hacerlo repetible
x = np.sort(10 * np.random.rand(n, 1) - 5, axis=0)
y = 0.2 * x**3 - 0.5 * x + 2 + np.random.randn(n, 1)

#Añadir componente polinómica (x³) para el modelo polinómico
x_poli = np.hstack([x, x**3])

#Crear y entrenar modelo lineal
rl = LinearRegression().fit(x, y)
r2_lineal = rl.score(x, y)

#Crear y entrenar modelo polinómico
rp = LinearRegression().fit(x_poli, y)
r2_polinomico = rp.score(x_poli, y)

#Mostrar R² de cada modelo
print(f"R² del modelo lineal: {r2_lineal:.4f}")
print(f"R² del modelo polinómico: {r2_polinomico:.4f}")

#Crear datos de prueba para graficar las predicciones
rejilla_lineal = np.linspace(-5, 5, 100).reshape(-1, 1)
rejilla_polino = np.hstack([rejilla_lineal, rejilla_lineal**3])

#Graficar los resultados
plt.scatter(x, y, color="black", label="Datos")
plt.plot(rejilla_lineal, rl.predict(rejilla_lineal), color="blue", label="Lineal")
plt.plot(rejilla_lineal, rp.predict(rejilla_polino), color="red", label="Polinómica")
plt.legend()
plt.title("Comparación de modelo lineal vs polinómico")
plt.show()
```

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Crear semilla personalizada basada en el nombre completo
nombre = "Juan"
ape1 = "Martínez"
ape2 = "Pérez"
semilla = sum(ord(c) for c in (nombre + ape1 + ape2))
print(f"Semilla personalizada: {semilla}")

#Generar datos
n = 100 # Número de datos
np.random.seed(semilla) # Usar la semilla personalizada
x = np.sort(10 * np.random.rand(n, 1) - 5, axis=0)
y = 0.2 * x**3 - 0.5 * x + 2 + np.random.randn(n, 1)

#Añadir componente polinómica (x³) para el modelo polinómico
x_poli = np.hstack([x, x**3])

#Dividir en datos de entrenamiento y prueba (80% train, 20% test) usando la semilla personalizada
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=semilla)
x_poli_train, x_poli_test = train_test_split(x_poli, test_size=0.2, random_state=semilla)

#Crear y entrenar modelo lineal
rl = LinearRegression().fit(x_train, y_train)
r2_lineal_train = rl.score(x_train, y_train)
r2_lineal_test = rl.score(x_test, y_test)

#Crear y entrenar modelo polinómico
rp = LinearRegression().fit(x_poli_train, y_train)
r2_polinomico_train = rp.score(x_poli_train, y_train)
r2_polinomico_test = rp.score(x_poli_test, y_test)

#Mostrar R² de cada modelo
print(f"R² del modelo lineal (train): {r2_lineal_train:.4f}")
print(f"R² del modelo lineal (test): {r2_lineal_test:.4f}")
print(f"R² del modelo polinómico (train): {r2_polinomico_train:.4f}")
print(f"R² del modelo polinómico (test): {r2_polinomico_test:.4f}")

#Crear datos de prueba para graficar las predicciones
rejilla_lineal = np.linspace(-5, 5, 100).reshape(-1, 1)
rejilla_polino = np.hstack([rejilla_lineal, rejilla_lineal**3])

#Graficar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color="black", label="Datos (train)")
plt.scatter(x_test, y_test, color="gray", label="Datos (test)")
plt.plot(rejilla_lineal, rl.predict(rejilla_lineal), color="blue", label="Lineal")
plt.plot(rejilla_lineal, rp.predict(rejilla_polino), color="red", label="Polinómica")
plt.legend()
plt.title("Comparación de modelo lineal vs polinómico (train y test)")
plt.show()
```

Tenemos los datos de 'Grade_Set_2.csv', considera otro conjunto de datos de notas de estudiantes

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#Carga del csv
df = pd.read_csv('Grade_Set_2.csv')
print(df)

#Gráfico
df.plot(kind='scatter', x='Hours_Studied', y='Test_Grade', title='Grade vs Hours Studied')
plt.show() # added to show the graph.

print("Correlation Matrix: ")
print(df.corr()) # print the correlation matrix

#Create linear regression object
lr = LinearRegression()
x = df.Hours_Studied.values.reshape(-1, 1)  #independent variable, reshape is required
y = df.Test_Grade  # dependent variable

#Entrenamos el modelo usando los sets de train
lr.fit(x, y)
plt.scatter(x, y, color='black')
plt.plot(x, lr.predict(x), color='blue', linewidth=3)
plt.title('Grade vs Hours Studied')
plt.ylabel('Test_Grade')
plt.xlabel('Hours_Studied')
plt.show() # added to show the graph

print("R Squared: ", r2_score(y, lr.predict(x)))
```

El análisis de correlación muestra el 97% de relación positiva entre las horas estudiadas y la nota obtenida y el 95% (R2) de variación puede ser explicada por las horas estudiadas. Observa que menos
de 4 horas de media estudiadas da una puntuación menor de 30 y cada 9 horas añade a la nota un nivel. Esto puede modelarse también con un regresor polinómico.

```python
lr = LinearRegression()
x= df.Hours_Studied # independent variable
y= df.Test_Grade # dependent variable
for deg in [1, 2, 3, 4, 5]:
 lr.fit(np.vander(x, deg + 1), y) # función vander() devuelve potencias del vector
 y_lr = lr.predict(np.vander(x, deg + 1))
 plt.plot(x, y_lr, label='degree ' + str(deg))
 plt.legend(loc=2)
 print (r2_score(y, y_lr))
 plt.plot(x, y, 'ok')
```


**d) Modifica el meta-modelo soft-voting pasándole esos 5 modelos base, y además le pasas los pesos modificados de manera que el peso del modelo i-ésimo sea su accuracy dividido por la suma de todas las accuracies**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, VotingClassifier
from sklearn import metrics

#Carga de los datos
df = pd.read_csv("Diabetes.csv")
X = df[['age', 'serum_insulin']]  # variables independientes
y = df['class'].values  # variable dependiente

#Normalizamos los datos
X = StandardScaler().fit_transform(X)

#Dividimos los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2017)

#Inicializamos los clasificadores base
models = [
    ('Decision Tree', DecisionTreeClassifier(max_depth=1, random_state=2017)),
    ('KNN', KNeighborsClassifier()),
    ('SVM', SVC(probability=True)),  # Habilitamos la probabilidad para SVM
    ('Logistic Regression', LogisticRegression()),
    ('Random Forest', RandomForestClassifier(n_estimators=100, random_state=2017)),
    ('AdaBoost (Decision Tree)', AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1, random_state=2017), n_estimators=100, learning_rate=0.1, random_state=2017))
]

#Calculamos el accuracy de cada modelo utilizando validación cruzada
kf = StratifiedKFold(n_splits=5, random_state=2017, shuffle=True)
accuracies = {}

for name, model in models:
    cv_results = cross_val_score(model, X_train, y_train, cv=kf, scoring='accuracy')
    accuracies[name] = cv_results.mean()

#Cálculo de la suma total de las precisiones
total_accuracy = sum(accuracies.values())

#Cálculo de los pesos de cada modelo
weights = {name: accuracy / total_accuracy for name, accuracy in accuracies.items()}

#Creamos el meta-modelo con soft voting (votación ponderada)
voting_model = VotingClassifier(estimators=models, voting='soft', weights=[weights[name] for name, _ in models])

#Entrenamos el meta-modelo
voting_model.fit(X_train, y_train)

#Evaluamos el modelo con validación cruzada (5 pliegues)
cv_results = cross_val_score(voting_model, X_train, y_train, cv=kf, scoring='accuracy')

#Imprimos el resultado de la validación cruzada
print(f"Soft Voting (Meta-modelo) - Train Accuracy: {cv_results.mean():.4f}")

#Evaluamos en el conjunto de prueba
y_pred = voting_model.predict(X_test)
test_accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Soft Voting (Meta-modelo) - Test Accuracy: {test_accuracy:.4f}")

#Mostramos los pesos de los modelos
print("\nPesos de los modelos en el meta-modelo Soft Voting:")
for name, weight in weights.items():
    print(f"{name}: {weight:.4f}")
```
