Como todo es un proceso aleatorio, para poder obtener resultados comparables vamos a utilizar la

Haz un gráfico scatterplot() de los datos, con los datos de entrenamiento como círculos de color rojo
y los de test círculos de color verde. Pon leyenda a cada variable y título "Inversión I+D vs Beneficio
de <tus iniciales>".

```python
#Carga de Librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

```python
#Carga del csv
datos = pd.read_csv('50_Startups.csv')
```

Haz un gráfico scatterplot() de los datos, con los datos de entrenamiento como círculos de color rojo
y los de test círculos de color verde. Pon leyenda a cada variable y título "Inversión I+D vs Beneficio
de <tus iniciales>"

```python
#Gráfico Scatterplot
#Usar columnas relevantes
X = datos['R&D Spend']
y = datos['Profit']

#Dividir los datos en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

#Crear scatterplot
plt.figure(figsize=(8, 5))
plt.scatter(X_train, y_train, color='red', label='Train')
plt.scatter(X_test, y_test, color='green', label='Test')
plt.xlabel('Inversión en I+D (R&D Spend)')
plt.ylabel('Beneficio (Profit)')
plt.title('Inversión I+D vs Beneficio de Juan')
plt.legend()
plt.show()
```

Antes del código que hace el gráfico crea un objeto LinearRegresion() y lo entrenas. Añade al gráfico
la línea recta de predicción a todos los datos de test en color azul

```python
#Gráfico Scatterplot
#Usar columnas relevantes


X = datos[['R&D Spend']]  #Lleva dos corchetes porque x debe ser dataframe y no una serie (2D)
y = datos['Profit']

#Dividir los datos en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Predecir con los datos de test
y_pred = model.predict(X_test)

#Crear scatterplot
plt.figure(figsize=(8, 5))
plt.scatter(X_train, y_train, color='red', label='Train')
plt.scatter(X_test, y_test, color='green', label='Test')
plt.plot(X_test, y_pred, color='blue', linewidth=2, label='Predicción')
plt.xlabel('Inversión en I+D (R&D Spend)')
plt.ylabel('Beneficio (Profit)')
plt.title('Inversión I+D vs Beneficio de Juan')
plt.legend()
plt.show()
```
