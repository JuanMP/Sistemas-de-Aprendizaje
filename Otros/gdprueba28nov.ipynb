```python
import numpy as np
import matplotlib.pyplot as plt

# Crear los datos
X = np.linspace(-5, 5, 100)
Y = 3 * X**3 - 2 * X + 4
DerivadaY = 9 * X**2 - 2  

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función
ax.plot(X, Y, color='red', label='Y = 3X³ - 2X + 4')

# Graficar la derivada
ax.plot(X, DerivadaY, color='blue', label="Derivada Y' = 9X² - 2")

# Añadir etiquetas y leyenda
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend()

# Mostrar la gráfica
plt.show()
```

```python
X,Y = np.meshgrid(
    np.linspace(-5,5,100),
    np.linspace(-5,5,100)
    
)

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection="3d")
Z = 2 * X**2 - X * Y + Y

DerivadaX = 4 * X - Y
DerivadaY = -X + 1
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, DerivadaX, color="red")
ax.plot_surface(X, Y, DerivadaY, color="yellow")
```

```python
X,Y = np.meshgrid(
    np.linspace(-5,5,100),
    np.linspace(-5,5,100)
    
)

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection="3d")
Z = 2 * X**2 - X * Y + Y

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R) / R
ax.plot_surface(X,Y,Z, cmap="Spectral_r")
```
